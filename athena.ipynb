```python
from nltk.stem import WordNetLemmatizer
import nltk
from collections import defaultdict
```

```python
def freq_stats_corpora_unique(dictionnary, stopwords=None, len_words=2):
    """
    Lower, exclude stopwords, exclude words smaller than 3 letters, lemmanize
    
    Positionnal argument:
    ---------------------------
    dictionnary : dict: keys=identity, values=text to process
    
    Opt args :
    ---------------------------
    stopwords : list : list of words to exclude
    len_words : int : length mini of a word to be kept
    
    Return:
    ---------------------------
    freq = {clefs de dictionnary: {tokens : fréquence}}
    stats = {clefs de dictionnary: {'total' : nombre de token unique}}
    corpora = {clefs de dictionnary: {'total' : [tokens]}}
    """
    lemmatizer = WordNetLemmatizer()
    tokenizer = nltk.RegexpTokenizer(r'\w+')
    corpora = defaultdict(list)
    # Création d'un corpus de tokens par id
    if stopwords==None:
        stopwords = []
    stopwords = [lemmatizer.lemmatize(w) for w in stopwords]
    for identity,text in dictionnary.items():
        tokens  = tokenizer.tokenize(text)
        corpora[identity] += [lemmatizer.lemmatize(w) for w in tokens if not lemmatizer.lemmatize(w) in stopwords]
        corpora[identity] = [w for w in corpora[identity] if len(w)>len_words]
    stats, freq = dict(), dict()

    for k, v in corpora.items():
        freq[k] = nltk.FreqDist(v)
        stats[k] = len(freq[k])
        
    return (freq, stats, corpora)
```

```python
def freq_tot(corpora, freq):
    freq_tot = nltk.Counter()
    for k in corpora.keys():
        freq_tot += freq[k]
    freq_tot = pd.Series(freq_tot).sort_values(ascending=False)
    freq_tot = freq_tot.to_dict()
    return freq_tot
```
